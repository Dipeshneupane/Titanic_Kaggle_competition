```python
def add(var):
    Age= var[0]
    Pclass = var[1]
    if pd.isnull(Age):
        if Pclass==1:
            return 38
        elif Pclass==2:
            return 30
        else:
            return 25
    else:
        return Age
```

```python
def age_group(age):
    a = ''
    if (age<=1):
        a = "infant"
    elif (age<=3):
        a = "toddler"
    elif (age<=12):
        a = "child"
    elif (age<=19):
        a = "teenager"
    elif (age<=25):
        a = "young adult"
    elif (age<=40):
        a = "adult"
    elif (age<=55):
        a = 'middle age'
    else:
        a = "old"
    return a
```
